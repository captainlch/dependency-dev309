```python
mycountry = ['VNM']
peers = ['CHN', 'LAO', 'THA', 'AGO', 'MMR', 'UZB', 'WSM']
```

## don't edit below here

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
mycountries = peers + mycountry
elec_out = pd.read_csv('elec_out.csv')
elec_out = elec_out.drop('Unnamed: 0', axis=1)
water_out = pd.read_csv('water_out.csv')
water_out = water_out.drop('Unnamed: 0', axis=1)
rail_out = pd.read_csv('rail_out.csv')
rail_out = rail_out.drop('Unnamed: 0', axis=1)
truck_out = pd.read_csv('truck_out.csv')
truck_out = truck_out.drop('Unnamed: 0', axis=1)
internet_out = pd.read_csv('internet_out.csv')
internet_out = internet_out.drop('Unnamed: 0', axis=1)
contract_out = pd.read_csv('contract_out.csv')
contract_out = contract_out.drop('Unnamed: 0', axis=1)
ed_out = pd.read_csv('ed_out.csv')
ed_out = ed_out.drop('Unnamed: 0', axis=1)
hc_out = pd.read_csv('hc_out.csv')
hc_out = hc_out.drop('Unnamed: 0', axis=1)
capital_out = pd.read_csv('capital_out.csv')
capital_out = capital_out.drop('Unnamed: 0', axis=1)
land_out = pd.read_csv('land_out.csv')
land_out = land_out.drop('Unnamed: 0', axis=1)
```

## graphs are down here

```python
#electricity
mycvar = 'elec_' + mycountry[0]
mycvars = []
for i in mycountries:
    mycvars.append('elec_' + i)

fig, ax = plt.subplots(figsize=(5,5))
ax.scatter(elec_out.loc[~elec_out.ID.isin(mycvars)]['year'], elec_out.loc[~elec_out.ID.isin(mycvars)]['mean'], c="gray")
ax.scatter(elec_out.loc[elec_out.ID.isin(mycvars)]['year'], elec_out.loc[elec_out.ID.isin(mycvars)]['mean'], c="blue")
ax.scatter(elec_out.loc[elec_out.ID == mycvar]['year'], elec_out.loc[elec_out.ID == mycvar]['mean'], c="red")
(_, caps, _) = ax.errorbar(elec_out.loc[elec_out.ID == mycvar]['year'], elec_out.loc[elec_out.ID == mycvar]['mean'], yerr=elec_out.loc[elec_out.ID == mycvar]['error'], fmt='o', capsize=5, c="red")
for cap in caps:
    cap.set_markeredgewidth(1)
ax.legend(['Rest of World', 'Peers', mycountry[0]], loc=9, bbox_to_anchor=(0.5, -0.15), ncol=2)
plt.xlabel('Year', fontsize=12)
plt.ylabel('Dependency Coefficient', fontsize=12)
plt.title('Concentration of Exports by Electricity Dependence, 1995 - 2015')
del mycvar
del mycvars
plt.show()
```
